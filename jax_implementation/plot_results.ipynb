```python
import pandas as pd
```

```python
cifar1 = pd.read_csv('cifar_1_2.csv')
cifar1 = cifar1[cifar1['test_step'] == 3359]

# Calculate mean accuracy, std dev and counts in one go
aggregated_stats = cifar1.groupby(['learning_rate', 'weight_decay', 'adam_corr', 'eps_root']).agg(
    mean_accuracy=('best_test_accuracy', 'mean'),
    std_accuracy=('best_test_accuracy', 'std'),
    count=('best_test_accuracy', 'size')
).reset_index()

# Sort by accuracy
aggregated_stats = aggregated_stats.sort_values('mean_accuracy', ascending=False)

print("\nAggregated statistics (accuracy, std dev and counts):")
print(aggregated_stats)
```

```python
# find the top 10 results for adam_corr = True
top_10_adam_corr_true = aggregated_stats[aggregated_stats['adam_corr'] == True].head(10)
print("\nTop 10 results for adam_corr = True:")
print(top_10_adam_corr_true)

# find the top 10 results for adam_corr = False
top_10_adam_corr_false = aggregated_stats[aggregated_stats['adam_corr'] == False].head(10)
print("\nTop 10 results for adam_corr = False:")
print(top_10_adam_corr_false)
```
